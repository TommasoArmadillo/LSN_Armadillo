```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8, 5]

x, y, error = np.loadtxt("Put_direct.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error ,fmt='orange')
plt.xlabel('#throws')
plt.ylabel('Put_value')
plt.grid(True)
plt.plot([0, 10000], [5.4595325819072364, 5.4595325819072364], color='k', linestyle='-', linewidth=2)
plt.title("Put option one step")
axes = plt.gca()
axes.set_ylim([4,6.5])
plt.show()
```
