```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y, error = np.loadtxt("output.ene.0", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error, linewidth = 0., marker = '.', ecolor = 'black', elinewidth = 3., color = 'black')
plt.grid(True)
print(error)
points=100
T = np.linspace(0.2,3.0,num=points)
beta = 1/T
J = 1.0
Ns = 50
th = np.tanh(J/T)
thN= th**Ns
ch = 1/th
e = -J*( th + ch*thN )/( 1 + thN )
plt.plot(T, e, color = 'orange')
plt.title('Ising 1D, internal energy')
plt.xlabel('T')
plt.ylabel('U/N')

plt.show()
```
