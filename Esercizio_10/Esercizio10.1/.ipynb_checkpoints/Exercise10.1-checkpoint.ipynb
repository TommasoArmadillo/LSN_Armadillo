```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt("cities.final", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y, 'ro',linestyle='-', lw=0.4)
plt.show()
```
