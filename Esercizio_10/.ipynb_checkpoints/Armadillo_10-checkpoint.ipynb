## <span style="color:blue"> Exercise 10.1</span>
#### Scope
The scope of the exercise was to so solve the travelling salesman problem using a simulated annealing algorithm.

#### Instruction to compile
Code is in the folder <font face = "Courier"> Esercizio10.1 </font>. It can be compiled using <font face = "Courier"> `$make` </font> and executed with <font face = "Courier"> `$./main.exe` </font>. 
It reads the cities in input in the file 'cities.0' (if there isn't this file it generates some random cities) and produces two files as output: the lenght of the shortest path as a function of generations and the best path of the last generation.

#### Cities in circle
We gradually lowered the temperature from a starting value of $T=1$ to $T=0.001$ in step of $0.001$. For each step we moved the system 15 000 times using the Metropolis algorithm. 

We tried different solutions, for example starting from $T=10$ to $T=1$ in step of $1$ and then to $T=0.001$ in smaller steps, but the one we used is, so far, the easier to implement and the best one in  efficiency, because if $T=1$ the acceptance rate is $0.9$ so the system is as free to move as if the temperature would be higher.

Like the previous numerical exercise (Genetic algorithm) we started from a situation in which the cities are disposed in a circle, so it is easy to see if the algorithm works.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt("Esercizio10.1/cities.round.final", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y, 'ro',linestyle='-', lw=0.4)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y = np.loadtxt("Esercizio10.1/SmallestPath.round.dat", usecols=(0, 1), delimiter=' ', unpack='true')
plt.plot(x, y)
plt.xlabel('iterations')
plt.ylabel('SmallestPath')
plt.grid(True)
plt.show()
```

The path we found is the best one, so the algorithm is working properly.

#### Cities inside a square
We used the same configuration of the cities as the previous numerical exercise so we can compare results:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1,2)
plt.rcParams['figure.figsize'] = [15, 5]

x_SA, y_SA = np.loadtxt("Esercizio10.1/cities.square.final", usecols=(0,1), delimiter=' ', unpack='true')
x_GA, y_GA = np.loadtxt("../Esercizio_9/Esercizio9.1/cities.square.final", usecols=(0,1), delimiter=' ', unpack='true')
ax[0].plot(x_SA,y_SA, 'ro',linestyle='-', lw=0.4, color='red')
ax[0].set_title('Simulated Annealing')

ax[1].plot(x_GA,y_GA, 'ro',linestyle='-', lw=0.4, color='blue')
ax[1].set_title('Genetic Algorithm')

plt.show()
```

They are different! So one of the two algorithm found a better path. Let's look at the lenght of the two paths:
- Simulated annealing $\to 4.77046$
- Genetic algorithm $\to 4.77596$

So the SA algorithm found a better path! This means that the GA we created can be improved, so for example we could increase the number of generations or implement new mutations or a new crossover system. 

Let's see also how change the graph of the lenght of the best path.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

fig, ax = plt.subplots(1,2)
plt.rcParams['figure.figsize'] = [15, 5]

x_SA, y_SA = np.loadtxt("Esercizio10.1/SmallestPath.square.dat", usecols=(0,1), delimiter=' ', unpack='true')
x_GA, y_GA = np.loadtxt("../Esercizio_9/Esercizio9.1/SmallestPath.square.dat", usecols=(0,1), delimiter=' ', unpack='true')
ax[0].plot(x_SA,y_SA, 'ro',linestyle='-', lw=0.4, color='red')
ax[0].set_title('Simulated Annealing - best path')
ax[0].set_xlabel('beta')
ax[0].set_ylabel('lenght')

ax[1].plot(x_GA,y_GA, 'ro',linestyle='-', lw=0.4, color='blue')
ax[1].set_title('Genetic Algorithm - best path')
ax[1].set_xlabel('generation')
ax[1].set_ylabel('lenght')

plt.show()
```

We see that the simulated annealing start finding good paths only after many moves (for every value of beta we made 15 000 moves), while the genetic algorithm after 200-300 generations proposes a path which does not differ too much from the best one.

#### Time of execution
The two algorithm require more or less the same time to be executed. If we would like to increment the number of cities then the simulated annealing would be better. In fact the GA works with a population that grows as $n^2$ while the SA works with only one path that moves inside the space of all the possible paths.

## <span style="color:blue"> Exercise 10.1</span>

We have to parallelize our code using the MPI libraries to perform a Random Search with parallel SA searches of the optimal path: each node should perform an independent SA search and only in the end we will compare the results of each node. 

Unfortunately my laptop has only two cores, so we don't expect to observe a significant improvement.

#### Instruction to compile
Code is in the folder <font face = "Courier"> Esercizio10.2 </font>. It can be compiled using <font face = "Courier"> `$make` </font> and executed with <font face = "Courier"> `$mpiexec -np 2 main.exe` </font>. 
It reads the cities in input in the file 'cities.0' (if there isn't this file it generates some random cities) and produces three files as output: the lenght of the shortest path as a function of beta (one for each thread) and the best path of the last generation.

#### Circle
I plot the lenght of the best path as function of $\beta$ for 30 cities arranged in circle and a picture for the best path compared with the one we got from the non-parallelized code. These are my results.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [15, 6]

x_0, y_0 = np.loadtxt("Esercizio10.2/Results/SmallestPath.round.thread0.dat", usecols=(0,1), delimiter=' ', unpack='true')
x_1, y_1 = np.loadtxt("Esercizio10.2/Results/SmallestPath.round.thread1.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x_0, y_0, linestyle='-', color='red', label='thread 0')
plt.plot(x_1, y_1, linestyle='-', color='blue', label='thread 1')
plt.title('Simulated Annealing - best path - parallelized code')
plt.xlabel('beta')
plt.ylabel('lenght')
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(1,2)
plt.rcParams['figure.figsize'] = [15, 6]

x_SA, y_SA = np.loadtxt("Esercizio10.1/cities.round.final", usecols=(0,1), delimiter=' ', unpack='true')
x_par, y_par = np.loadtxt("Esercizio10.2/cities.round.final", usecols=(0,1), delimiter=' ', unpack='true')
ax[0].plot(x_SA,y_SA, 'ro',linestyle='-', lw=0.4, color='red')
ax[0].set_title('Simulated Annealing - non parallel')

ax[1].plot(x_par,y_par, 'ro',linestyle='-', lw=0.4, color='blue')
ax[1].set_title('Genetic Algorithm - parallel')

plt.show()
```

We found the same result because this is already the best path.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [15, 6]

x_0, y_0 = np.loadtxt("Esercizio10.2/Results/SmallestPath.square.thread0.dat", usecols=(0,1), delimiter=' ', unpack='true')
x_1, y_1 = np.loadtxt("Esercizio10.2/Results/SmallestPath.square.thread1.dat", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x_0, y_0, linestyle='-', color='red', label='thread 0')
plt.plot(x_1, y_1, linestyle='-', color='blue', label='thread 1')
plt.title('Simulated Annealing - best path - parallelized code')
plt.xlabel('beta')
plt.ylabel('lenght')
plt.legend()

plt.show()
```

```python
fig, ax = plt.subplots(1,2)
plt.rcParams['figure.figsize'] = [15, 6]

x_SA, y_SA = np.loadtxt("Esercizio10.1/cities.square.final", usecols=(0,1), delimiter=' ', unpack='true')
x_par, y_par = np.loadtxt("Esercizio10.2/cities.square.final", usecols=(0,1), delimiter=' ', unpack='true')
ax[0].plot(x_SA,y_SA, 'ro',linestyle='-', lw=0.4, color='red')
ax[0].set_title('Simulated Annealing - non parallel')

ax[1].plot(x_par,y_par, 'ro',linestyle='-', lw=0.4, color='blue')
ax[1].set_title('Genetic Algorithm - parallel')

plt.show()
```
