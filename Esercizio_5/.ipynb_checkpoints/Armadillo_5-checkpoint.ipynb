#### Scope
The scope of the exercise was to use the Metropolis algorithm to sample $|\Psi_{1,0,0}(x,y,z)|^2$ and $|\Psi_{2,1,0}(x,y,z)|^2$ in Cartesian coordinates using an uniform transition probability $T(\vec{x}|\vec{y})$. #### $|\Psi_{1,0,0}(x,y,z)|^2$
This wave function represent the ground state for the hydrogen atom. It's explicit form in spherical coordinates is
$$
\Psi_{1,0,0}(r,\theta,\phi)=
\frac{a_0^{3/2}}{\sqrt{\pi}} e^{-r/a_0} 
$$
were $a_0$ is Bohr radius. To come back to cartesian coordinates we calculate $r=\sqrt{x^2+y^2+z^2}$ and $\theta=z/r$. We start from a arbitrary point in space and we let the Metropolis algorithm equilibrate. So if we start from a point very far from where the probability density is significally not null, it will reach on his own a higher probability region and the points that starts generating from this moment on, better describe the probability density we are sampling from. We can then modify the value of $\delta$ in order to make the acceptance rate to hit the 'law of $50%$'. We work in Bohr radius units, so $r \to r/a_0$, and we can take $a_0=1$.

For each step we calculated the value of $|\vec{r}|=r$ and we computed it's mean value through block averaging. We sampled $M=10^6$ points divided in 1 000 blocks, so every block contains 1 000 values for r. I think we could have used very less values in each block, maybe even 100, because $\delta=1.2$, and the $|\Psi_{1,0,0}|^2$ distribution has a spherical form with mean radius (from exact calculation)
$$ \langle r \rangle_{\Psi_{1,0,0}}=\frac{3}{2}$$
so points get scorrelated very soon.

We started from the point $\vec{x_0}=(0,0,0)$ because it's the centre of the sphere, where the probability is maximum, in this way we require less step to equlibrate. Otherwise, if we would have started from another point, for example $\vec{x_0}=(5,0,0)$ we should have waited more steps ($\sim 40$ steps).

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8, 5]

x, y, error = np.loadtxt("Esercizio5.1/Radius100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error ,fmt='orange')
plt.xlabel('#throws')
plt.ylabel('Radius')
plt.grid(True)
plt.title("Mean radius for Psi100")
plt.plot([0, 100000], [1.5, 1.5], color='k', linestyle='-', linewidth=2)
plt.show()
```

The mean value for $r$ is compatible whit exact calculation. The acceptance rate is 0.51024. Down here we plotted 10 000 points. We can see they clearly disposed in a sphere as we expected.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z = np.loadtxt("Esercizio5.1/Psi100.txt", usecols=(0,1,2), delimiter=' ', unpack='true', max_rows = 10000)

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```

#### $|\Psi_{2,1,0}(x,y,z)|^2$

The exact expression for this probability density is 
$$
\Psi_{2,1,0}(r,\theta,\phi)=
\frac{a_0^{5/2}}{8}\sqrt{\frac{2}{\pi}} r e^{-r/2a_0} \cos(\theta)
$$
In this case we started from the point $\vec{x_0}=(0,0,2.5)$ and we used $\delta=2.8$. We waited 100 steps before starting measuring the radius to equilibrate the system.

The mean value of radius should be
$$\langle r\rangle_{\Psi_{2,1,0}}= 5$$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8, 5]

x, y, error = np.loadtxt("Esercizio5.1/Radius210.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error ,fmt='orange')
plt.xlabel('#throws')
plt.ylabel('Radius')
plt.grid(True)
plt.title("Mean radius for Psi210")
plt.plot([0, 100000], [5, 5], color='k', linestyle='-', linewidth=2)
plt.show()
```

Mean value is compatible with the expected value. The acceptance rate is 0.52367.
