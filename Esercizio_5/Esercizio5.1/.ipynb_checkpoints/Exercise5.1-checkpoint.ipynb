```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [8, 5]

x, y, error = np.loadtxt("Radius100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error ,fmt='orange')
plt.xlabel('#throws')
plt.ylabel('Radius')
plt.grid(True)
plt.title("Mean radius for Psi100")
plt.plot([0, 10000], [1.5, 1.5], color='k', linestyle='-', linewidth=2)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

X, Y, Z = np.loadtxt("Psi100.txt", usecols=(0,1,2), delimiter=' ', unpack='true')

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter(X, Y, Z, c=Z, marker='.')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.view_init(10, 30)
plt.show()
```
