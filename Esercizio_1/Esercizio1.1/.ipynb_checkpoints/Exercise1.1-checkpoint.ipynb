The scope of this first exercise was to test the Pseudo-Random Number generator downloaded from the NSL Ariel web site by estimating $\langle r \rangle = \int_0^1 r dr = 1/2$ and $\sigma^2 = \int_0^1 (r-1/2)^2 dr = 1/12$ and make a picture of the estimation of $\langle r \rangle$ and $\langle (r-1/2)^2 \rangle$ along with their uncertainty (which corresponds to Standard Deviation of the mean) with a large number of *throws* $M$ (I used $M = 10^4$) as a function of the number of blocks, $N$ (i.e. using the <i>blocking method</i>).

I wrote a C++ program, which can be compiled using <font face = "Courier"> `make` </font> and executed with <font face = "Courier"> `./main.cpp` </font>that produces three files as output, one for each task of the exercise1.0.


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y, error = np.loadtxt("Mean.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-0.5,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<r>-0.5')
plt.grid(True)
plt.show()
```

This is the output for the mean value and Standard Deviation of the mean for the estimation of $\langle r \rangle$, we can clearly see that the average value of $\langle r \rangle - 1/2$ converges to 0 when <i>#throws</i> becomes bigger, and the error bar decrease in magnitude, as we expected. Actually 0 is not included in the error bar, but it's not a problem beacuse only in 68% of the cases it will end in the error bars, 95% of times will end in 2 error bar (which is our case). So in the end the value we found it's compatible with $\langle r \rangle = \int_0^1 r dr = 1/2$.

In order to calculate $\sigma^2$ I used the same code as the previous task and instead of generate random numbers between 0 and 1 and then calculate mean and Standard Deviation of the mean of this numbers, I generated numbers between 0 and 1 (uniformly distributed) and  calculate $s =  (r-1/2)^2$ and repeat the previous calculation (mean value and STD of mean) on $s$.

I produced a text file called <font face = "Courier"> `Variance.txt` </font> and the results are plot below:


```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y, error = np.loadtxt("Variance.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y-1/12,yerr=error)
plt.xlabel('#throws')
plt.ylabel('<(r-0.5)^2>-1/12')
plt.grid(True)
plt.show()
```

We can see that the average value of $\langle (r-1/2)^2 \rangle - 1/12$ converges to 0 when <i>#throws</i> becomes bigger, and the error bar decrease in magnitude, as we expected. This time 0 is included in the error bar, so we conclude that $\langle s \rangle = \langle (r-1/2)^2 \rangle = 1/12$ is compatible with what we estimated using Monte Carlo Methods.

In the third parte of the exercise we had to generate $n=10^4$ random numbers between 0 and 1, then divide the interval $\lbrack0,1\rbrack$ in $M=10^2$ sub-intervals and count how many random numbers there were in every interval. So we calculate the value for $\chi^2$: 
$$\chi^2 = \sum_{i=1}^M \frac{\left( n_i - n/M \right)^2}{n/M}$$
We should expect on average that $(n_i - n/M)^2 \simeq n/M$ and thus $\chi^2 \simeq 100$, i.e. the mean value for $\chi^2$ should be equal to the number of sub-intervals.
We repeated the calculation 100 times. The results are shown below:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y= np.loadtxt("CHI2.txt", usecols=(0,1), delimiter=' ', unpack='true')
plt.plot(x,y-100)
plt.xlabel('#test')
plt.ylabel('$\chi^2-100$')
plt.grid(True)
plt.show()

```

We see that the 100 values for $\chi^2-100$ we estimated distribute more or less uniformly around 0. So it's plausible that the numbers we generated comes from a uniform distribution between 0 and 1.
