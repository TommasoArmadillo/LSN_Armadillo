```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y, error = np.loadtxt("Uniform.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error)
plt.xlabel('#throws')
plt.ylabel('Integral')
plt.grid(True)
plt.show()
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x, y, error = np.loadtxt("Importance.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x,y,yerr=error)
plt.xlabel('#throws')
plt.ylabel('Integral')
plt.grid(True)
axes = plt.gca()
axes.set_ylim([0.940,1.105])
plt.show()
```
