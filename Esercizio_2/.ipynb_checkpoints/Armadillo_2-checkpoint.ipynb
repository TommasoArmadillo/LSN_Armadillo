## <span style="color:blue"> Exercise 2.1</span>
#### Scope
The scope of the exercise was to compute the integral
$$I = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) dx = 1$$
using a Monte Carlo method, both sampling a uniform distribution between 0 and 1 and using importance sampling.

#### Instruction to compile
Code is in the folder <font face = "Courier"> Esercizio1.1 </font>. It can be compiled using <font face = "Courier"> `$make` </font> and executed with <font face = "Courier"> `$./main.cpp` </font>. There is also the possibility to use <font face = "Courier"> `$make clean`</font>. It produces two files as output: the first one (<i>'Uniform.txt'</i>) contains the mean value of the integral with is uncertainty as a function of the number of blocks, the second one (<i>'Importance.txt'</i>) contains the mean value of the integral calcultated with importance sampling.

#### Uniform distribution

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x1, y1, error1 = np.loadtxt("Esercizio2.1/Uniform.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x1,y1,yerr=error1)
plt.xlabel('#throws')
plt.ylabel('Integral')
plt.title('Integral calculated with uniform distribution')
plt.grid(True)
plt.show()
```

The value of $I$ approches 1 as expected.

#### Importance sampling
To use the method of the importance sampling we rewrite the integral as 
$$I = \int_{0}^{1} \frac{\pi}{2}\cos(\pi x/2) dx = \int_0^1 \frac{\pi}{2}\cos(\pi x/2) \frac{p(x)}{p(x)}dx$$
where $p(x)\ge0 \forall x\in [0,1] $ and $\int_{0}^{1}p(x)dx=1$. Instead of generating points uniformly distributed in $[0,1]$ we generate points according to $p(x)$. It can be shown that the more $p(x)$ approximate the integrand, the more we can reduce the variance. So we could expand the integrand with Taylor to the second order, then we could use the method of inversion of the cumulative to generate points distributed with $p(x)$. This approach is very complicated because the first non-zero terms in the expansion of cosine is $O(x^2)$, so when we calculate the cumulative we obtain a third degree polynomials which can't be inverted analitically. So we approximate the cosine with a linear function. We impose $p(0)=\frac{\pi}{2}\cos(\pi 0/2)=\frac{\pi}{2}$ and $p(1)=0$ $\to$ $p(x)=\frac{\pi}{2}(1-x)$, but it has to be normalized so $p(x) = 2(1-x)$. The cumulative is

$$F(x)=\int_{0}^{x} p(t)dt=\int_{0}^{x} 2(1-t)dt=2\big(t-\frac{t^2}{2}\big)\big|_{t=0}^{t=x}=2\big(x-\frac{x^2}{2}\big)$$

which can be easily inverted

$$ 2\big(x-\frac{x^2}{2}\big) =  y $$

$$ x^2 -2x +y =0$$

$$x=1\pm\sqrt{1-y}=x=1-\sqrt{1-y}$$

We take the minus because we wanted a number in $[0,1]$. In the end to calculate the value of $I$ we compute the mean value of $\frac{\pi}{2}\frac{\cos(\pi x/2)}{2(1-x)}$ generating number distributed according to $p(x)=2(1-x)$. The result are shown below

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

x2, y2, error2 = np.loadtxt("Esercizio2.1/Importance.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(x2,y2,yerr=error2)
plt.xlabel('#throws')
plt.ylabel('Integral')
plt.title('Importance sampling')
plt.grid(True)

plt.show()
```

We obtained a mean value for $I$ of 1, as the previous case. The difference is in the error bars that are 4 times smaller. We can truly appreciate this difference setting the y-axes bounds as the previous graphic.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.errorbar(x1,y1,yerr=error1)
plt.errorbar(x2,y2,yerr=error2, color='orange')
plt.xlabel('#throws')
plt.ylabel('Integral')
plt.grid(True)
plt.title('Comparison')
axes = plt.gca()
axes.set_ylim([0.940,1.105])
plt.show()
```

In this graph we can appreciate the improvement in the error that we have using the method of importance sampling.

## <span style="color:blue"> Exercise 2.2</span>
#### Scope
The scope of the exercise was to simulate $N=10^4$ 3D random walk starting from the origin, both in a cubic lattice with lattice constant $a=1$ and in the continuum, and grap the value of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ with his uncertainty as a function of the number of steps.

#### Instruction to compile
Code is in the folder <font face = "Courier"> Esercizio2.2 </font>. It can be compiled using <font face = "Courier"> `$make` </font> and executed with <font face = "Courier"> `$./main.cpp` </font>. There is also the possibility to use <font face = "Courier"> `$make clean`</font>. It produces three files as output: <i>'Discrete.txt'</i> and <i>'Continuum.txt'</i> that contain requested data in the two fifferent cases, and <i>'Example.txt'</i> that contains three examples of random walk.

#### Discrete case
In every step we make a step in a random direction (x, y, z) with a $50%$ chance forward or backward. The result are plotted below.

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

plt.rcParams['figure.figsize'] = [8, 5]

xD, yD, errorD = np.loadtxt("Esercizio2.2/Discrete.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(xD,yD,yerr=errorD)
plt.xlabel('#steps')
plt.ylabel("$sqrt(<r>^2)$")
plt.grid(True)
plt.show()

#try to fit f(N)=k sqrt(N)
def f (x, k, a):
    return k * np.sqrt(x) + a

popt, pcov = curve_fit(f, xD, yD)
plt.plot(xD, f(xD, *popt), 'r', zorder=2)

print('Best fit is f(N) = ',popt[0], ' * sqrt(N) + ', popt[1] )
```

A random walk in space looks like this:

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

plt.rcParams['figure.figsize'] = [8, 5]

x1, y1, z1 = np.loadtxt("Esercizio2.2/Example.txt", usecols=(0,1,2), delimiter=' ', unpack='true', skiprows = 0, max_rows = 100)
x2, y2, z2 = np.loadtxt("Esercizio2.2/Example.txt", usecols=(0,1,2), delimiter=' ', unpack='true', skiprows = 100, max_rows = 100)
x3, y3, z3 = np.loadtxt("Esercizio2.2/Example.txt", usecols=(0,1,2), delimiter=' ', unpack='true', skiprows = 200, max_rows = 100)

fig = plt.figure()
ax = Axes3D(fig)

ax.plot(x1,y1,z1)
ax.plot(x2,y2,z2)
ax.plot(x3,y3,z3)
plt.title('3 random walk')
plt.show()
```

We see that the average distance from the origin increase as the simulation proceed and so does the uncertainty. The value of $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ fits very well with a function $f(N)=a\sqrt{N}+b$ (red line).

#### Continuum case
To generate a random direction we have to generate random value for $\theta \in [0,\pi]$ and $\phi \in [0,2\pi]$ distributed accrodingly to the jacobian $\sin \theta $. So $\phi$ is generated from a uniform distribution in $[0,2\pi]$ and $\theta$ from a $\frac{1}{2}\sin\theta$ distribution in $[0,\pi]$ were the factor $\frac{1}{2}$ was introduced as a normalization. To do so, we use the method of inversion of the cumulative.

$$F(\theta)=\int_{0}^{\theta}\frac{1}{2} \sin \alpha d\alpha = -\frac{1}{2}\cos\alpha\big|_{\alpha=0}^{\theta}=\frac{1-\cos\theta}{2}=y$$

$$\theta = \arccos(1-2y)$$

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit

plt.rcParams['figure.figsize'] = [8, 5]

xC, yC, errorC = np.loadtxt("Esercizio2.2/Continuum.txt", usecols=(0,1,2), delimiter=' ', unpack='true')
plt.errorbar(xC,yC,yerr=errorC, zorder=1)
plt.xlabel('#steps')
plt.ylabel("$sqrt(<r>^2)$")
plt.grid(True)
plt.show()

#try to fit f(N)=k sqrt(N)
def f (x, k, a):
    return k * np.sqrt(x) + a

popt, pcov = curve_fit(f, xC, yC)
plt.plot(xC, f(xC, *popt), 'r', zorder=2)

print('Best fit is f(N) = ',popt[0], ' * sqrt(N) + ', popt[1] )
```

Also in this case we found that the average value for $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}$ inscrease as $f(N)=a\sqrt{N}+b$. This fact indicates that random walks follow a diffusive behaviour. In a diffusive process, in fact, the probability density of finding the particle in the position (x, y, z, t) satisfay:

$$ \frac{\partial p}{\partial t}= D \bigg(\frac{\partial^2 p}{\partial x^2}+\frac{\partial^2 p}{\partial y^2}+\frac{\partial^2 p}{\partial z^2}\bigg)$$

So the solution is 

$$  p(x,y,z,t) = \frac{1}{\sqrt{4\pi D t}} e^{-\frac{x^2+y^2+z^2}{4Dt}}$$

We can now calculate $\langle \vec{r}_N \rangle_{RW}=\vec{0}$ and $\langle |\vec{r}_N|^2 \rangle_{RW} = 2Dt$ and so and so $\sqrt{\langle |\vec{r}_N|^2 \rangle_{RW}}\propto \sqrt{t}$ as we observed.
