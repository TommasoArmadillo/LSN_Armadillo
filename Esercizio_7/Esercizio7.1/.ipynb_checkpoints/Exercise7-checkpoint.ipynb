```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np

plt.rcParams['figure.figsize'] = [10, 5]

En = np.loadtxt("Potentialenergy.dat", usecols=(0), unpack='true', skiprows = 112000)
Pr = np.loadtxt("Pression.dat", usecols=(0), unpack='true')
plt.plot(En)
plt.xlabel('MC time-step')
plt.ylabel('PotentialEnergy')
plt.grid(True)
plt.title("Potential Energy")
plt.show()

plt.plot(Pr, color='orange')
plt.xlabel('MC time-step')
plt.ylabel('Pression')
plt.grid(True)
plt.title("Pression")
plt.show()
```

```python
#AUTOCORRELATION FUNCTION

#Calculating it for time from 0 to 1000
energy_mean = 0
energy_variance = 0
pression_mean = 0
pression_variance = 0
for i in range (0,100000):
    energy_mean += En[i]
    energy_variance += En[i]**2
    pression_mean += Pr[i]
    pression_variance += Pr[i]**2
    
energy_mean /= 100000
pression_mean /= 100000

energy_variance /= 100000
pression_variance /= 100000

energy_variance = energy_variance - energy_mean**2
pression_variance = pression_variance - pression_mean**2

print(energy_variance)
print(pression_variance)
print(np.var(En))
print(np.var(Pr))
```

```python
AC_energy = []
for i in range(0,500):
    mean_tmp = 0.
    count = 1
    for j in range (0,100000-i):
        mean_tmp += En[j] * En[i+j]
        count+=1
    mean_tmp /= count
    AC_energy.append( (mean_tmp - energy_mean**2) / energy_variance )
        
plt.plot(AC_energy, color='blue')
plt.xlabel('MC time-step')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.title("Autocorrelation Energy")
plt.show()       
```

```python
from scipy.optimize import curve_fit

x = []
for i in range(0,500):
    x.append(i)

#try to fit f(N)= EXP(-x/a)
def f (x, a):
    return np.exp(-x/a) 

popt, pcov = curve_fit(f, x, AC_energy)
plt.plot(x, AC_energy, color='blue')
x = np.asarray(x)
plt.plot(x, f(x, *popt), 'r')
plt.show()
print('Best fit is f(t) = exp(-t/', popt[0], ' )' )
```
